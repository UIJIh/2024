```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException, ElementClickInterceptedException, TimeoutException
import pandas as pd
import time

def click_element_with_retries(driver, element, max_retries=2):
    for _ in range(max_retries):
        try:
            element.click()
            return
        except StaleElementReferenceException:
            continue
    raise StaleElementReferenceException("Stale element reference: Element not found.")

data_list = []
end_flg = 0
start_time = time.time()

for i in range(10):
    if end_flg:
        break
        
    url_template = 'https://www.joongang.co.kr/generalelection2024/news?page={}'
    url = url_template.format(i+1)
    driver = webdriver.Chrome()
    driver.get(url)
    
    time.sleep(1)
    
    for n in range(24):    
        # click
        xpath_template = '//*[@id="news-container"]/div/div[1]/ul/li[{}]/div/h2/a'
        xpath = xpath_template.format(n+1)      
        try:
            main_title = WebDriverWait(driver, 3).until(EC.element_to_be_clickable((By.XPATH, xpath)))
            main_title.click()
            time.sleep(3)
        except:
            print(f"Failed to click. : {i}p,{n}th")
        
        # title
        title = WebDriverWait(driver, 3).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="container"]/section/article/header/h1')))
        title_text = title.text  
        
        # date
        try:
            date = WebDriverWait(driver, 3).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="container"]/section/article/header/div[2]/div/p[1]/time')))
            date_text = date.text 
            if "2024.03.31" in date_text:
                end_flg = True 
                print(f"Found article with date 2024.03.31, stopping...") 
                break
        except:
            date_text = None
        
        # update
        try :
            update = WebDriverWait(driver, 3).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="container"]/section/article/header/div[2]/div[1]/p[2]/time')))
            update_text = update.text 
        except: 
            update_text = None
            
        # journalist
        try:
            journalists_elements = WebDriverWait(driver, 4).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="container"]/section/article/header/div[3]//a')))
            journalists_text = ', '.join([element.text for element in journalists_elements])
        except:
            journalists_text = None
            print(f"Failed to find journalist. : {i}p,{n}th, {title_text}")
            
        # title2 
        try:
            t2 = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="article_"]/header/h2')))
            t2_text = t2.text
        except:
            t2_text = None
            
        # article
        try:
            paragraphs = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="article_body"]/p')))
            article_text = ', '.join([paragraph.text for paragraph in paragraphs])
        except TimeoutException:
            print(f"Failed to find article paragraphs within the given time. : {i}p,{n}th, {title_text}")
            
        data_list.append({
            'title': title_text,    
            'title2': t2_text,
            'date' : date_text,
            'update': update_text,
            'journalist': journalists_text,
            'article': article_text    
        })
    
        driver.back()
        time.sleep(3)
        
    driver.quit()
    time.sleep(1)
end_time = time.time()
total_time = end_time - start_time
print(f'Total time spent: {total_time}')
driver.quit()
```
