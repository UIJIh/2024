```python
# 1.
import numpy as np

# 최소자승법; 직선회귀식
def Linear_reg(x, y):
    n = len(x)
    
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xx = np.sum(x**2)
    sum_xy = np.sum(x*y)
    
    a0 = (sum_xx*sum_y - sum_xy*sum_x) / (n*sum_xx - sum_x**2)
    a1 = (n*sum_xy - sum_x*sum_y) / (n*sum_xx - sum_x**2)

    return a0, a1
```

```python
x = np.array([-1, 1, 2, 3, 7])
y = np.array([3, -1, 4, 2, 10])
s = Linear_reg(x, y)
```

```python
def Cubic_reg(x, y):
    n = len(x)
    sum_x = np.sum(x)
    sum_x2 = np.sum(x**2)
    sum_x3 = np.sum(x**3)
    sum_x4 = np.sum(x**4)
    sum_y = np.sum(y)
    sum_xy = np.sum(x*y)
    sum_x2y = np.sum((x**2)*y)
    A = np.array([[n, sum_x, sum_x2], 
                  [sum_x, sum_x2, sum_x3], 
                  [sum_x2, sum_x3, sum_x4]])
    b = np.array([sum_y, sum_xy, sum_x2y])
    x = np.linalg.solve(A, b)
    
    a0, a1, a2 = x
    #a0 = x[0], a1 = x[1], a2 = x[2]
    
    print('2차 다항식 회귀식 P(x) = %.4f + %.4f x + %.4f x^2' % (a0, a1, a2))
    
    return a0, a1, a2
```

```python
x = np.array([-1, 1, 2, 3, 7])
y = np.array([3, -1, 4, 2, 10])
s = Cubic_reg(x, y)
```

```python
# 2.
import numpy as np

def log_linear_reg(x, y):
    n = len(x)
    yy = np.log(y)
    
    sum_x = np.sum(x)
    sum_y = np.sum(yy)
    sum_xx = np.sum(x**2)
    sum_xy = np.sum(x * yy)
    
    a1 = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x**2)
    a0 = (sum_xx * sum_y - sum_xy * sum_x) / (n * sum_xx - sum_x**2)
    a0 = np.exp(a0)
    
    print(f'지수 함수 회귀식 p(x) = {a0} * e^({a1:.4f} * x)')
    return a0, a1

# 예제 데이터
x = np.array([1940, 1950, 1960, 1970, 1980, 1990, 2000])
y = np.array([537, 557, 682, 826, 981, 1135, 1262])

# 지수 함수 회귀 분석 수행
a0, a1 = log_linear_reg(x, y)

print("="*30)
# 주어진 데이터
years = np.array([1940, 1950, 1960, 1970, 1980, 1990, 2000])
population = np.array([537, 557, 682, 826, 981, 1135, 1262])

# 로그 변환
log_population = np.log(population)

# 선형 회귀 분석
coeffs = np.polyfit(years, log_population, 1)
b = coeffs[0]
log_a = coeffs[1]
a = np.exp(log_a)

# 지수 함수 회귀식
print(f'지수 함수 회귀식 P(t) = {a:.3f} * e^({b:.3f} * t)')

# 1955년의 인구 예측
year_to_predict = 1955
predicted_population = a * np.exp(b * year_to_predict)
print(f'1955년의 예상 인구: {predicted_population:.2f} 백만명')
```

```python
# 7.

# 미분방정식 (수직운동방정식의) 
def f(t, y):
    return 2 * t - y

# 초기값 y(0) = -1
y0 = -1

# 구간 [0, 1], 반복 회수 10회
t0 = 0
t_end = 1
N = 10

# 2차 Runge-Kutta 방법
h = (t_end - t0) / N
t = t0
y = y0
for i in range(N):
    k1 = f(t, y)
    k2 = h * f(t + h, y + k1)
    y = y + (k1 + k2) / 2
    t = t + h
print("2차 Runge-Kutta 방법으로 구한 y(1):", y)

# 4차 Runge-Kutta 방법
t = t0
y = y0
for i in range(N):
    k1 = h * f(t, y)
    k2 = h * f(t + h/2, y + k1/2)
    k3 = h * f(t + h/2, y + k2/2)
    k4 = h * f(t + h, y + k3)
    y = y + (k1 + 2*k2 + 2*k3 + k4) / 6
    t = t + h
print("4차 Runge-Kutta 방법으로 구한 y(1):", y)
```

```python
import sympy as sp

# Constants
g = 9.81  # Acceleration due to gravity in m/s^2
v0_horizontal = 10 * 1000 / 3600  # Initial horizontal velocity in m/s
y0 = 100  # Initial height in meters

# Calculate the vertical component of initial velocity
v0_vertical = 0  # Since the ball is thrown horizontally

# Define the variable
t = sp.symbols('t')

# Define the equation for vertical motion
y_t = y0 + v0_vertical * t - (1/2) * g * t**2

# Find the time when y(t) = 0
solution = sp.solve(y_t, t)

# Filter out negative solutions (assuming the positive solution is the time until it hits the ground)
positive_solutions = [sol.evalf() for sol in solution if sol > 0]

# Calculate the distance traveled horizontally
x_distance = v0_horizontal * positive_solutions[0]

# Print the solution
if len(positive_solutions) > 0:
    print("Time taken for the ball to hit the ground:", positive_solutions[0], "seconds")
    print("Distance traveled horizontally:", x_distance, "meters")
else:
    print("The ball does not hit the ground within the given initial conditions.")
```

1. 공이 100미터 높이에서 던져진다.
2. 초기 속도는 수평 방향으로 10km/s (즉, 10,000m/s)이다.
3. 중력 가속도 \( g = 9.81 \, \text{m/s}^2 \)이다.

우리는 공이 지면에 닿을 때까지 걸리는 시간을 계산할 것입니다.

### 수직 운동

수직 방향의 운동 방정식은 다음과 같습니다:
\[ y(t) = y_0 + v_{y0} t - \frac{1}{2} g t^2 \]

여기서:
- \( y_0 = 100 \, \text{m} \)
- 초기 수직 속도 \( v_{y0} = 0 \, \text{m/s} \) (수평으로 던져지므로)
- \( g = 9.81 \, \text{m/s}^2 \)

따라서:
\[ 0 = 100 - \frac{1}{2} \cdot 9.81 \cdot t^2 \]

이 방정식을 풀면:
\[ \frac{1}{2} \cdot 9.81 \cdot t^2 = 100 \]
\[ 4.905 \cdot t^2 = 100 \]
\[ t^2 = \frac{100}{4.905} \]
\[ t^2 \approx 20.387 \]
\[ t \approx 4.514 \, \text{seconds} \]

따라서 공이 지면에 도달하는 데 걸리는 시간은 약 4.514초입니다.

### 수평 운동

수평 방향의 운동은 등속 운동입니다. 수평 방향 초기 속도 \( v_{x0} = 10,000 \, \text{m/s} \) 이므로, 수평 거리 \( x \)는 다음과 같습니다:
\[ x = v_{x0} \cdot t \]

여기서 \( t \)는 우리가 구한 시간 4.514초입니다. 이 시간 동안 공이 이동한 수평 거리를 계산하면:
\[ x = 10,000 \cdot 4.514 \approx 45140 \, \text{meters} \]
