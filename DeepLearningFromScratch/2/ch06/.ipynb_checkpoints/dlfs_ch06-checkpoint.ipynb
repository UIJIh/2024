```python
import numpy as np
import matplotlib.pyplot as plt

N = 2
H = 3
T = 20

dh = np.ones((N ,H)) # 인수로 shape (N, H)
np.random.seed(3) # 난수 발생기의 시드 값을 고정시켜 실험의 재현성을 보장 (숫자에 따라 결과값 달라짐)
Wh = np.random.randn(H, H)

norm_list = []
for t in range(T):
    dh = np.matmul(dh, Wh.T)
    norm = np.sqrt(np.sum(dh**2)) / N
    norm_list.append(norm)

#print(norm_list)

plt.plot(np.arange(len(norm_list)), norm_list)
plt.xticks([0, 4, 9, 14, 19], [1, 5, 10, 15, 20]) # 눈금, 레이블기재
plt.xlabel('time step')
plt.ylabel('norm')
plt.show()
```

```python
import numpy as np

dW1 = np.random.randn(3, 3) * 10
dW2 = np.random.rands(3, 3) * 10
grads = [dW1, dW2]
max_norm = 5.0

def clip_grads(grads, max_norm):
    total_norm = 0
    for grad in grads:
        total_norm += np.sum(grad ** 2)
    total_norm = np.sqrt(total_norm)

    rate = max_norm / (total_norm + 1e-6)
    if rate < 1:
        for grad in grads:
            grad *= rate
```

```python
import sys
sys.path.append('..')
from common.functions import *

class LSTM:
    def __init__(self, Wx, Wh, b):
        self.params = [Wx, Wh, b]
        self.grads = [np.zeros_like(Wx), np.zeros_like(Wh), np.zeros_like(b)] # 모두 0으로!
        self.cache = None # 역전파 계산에 사용
    
    def forward(self, x, h_prev, c_prev):
        Wx, Wh, b = self.params
        N, H = h_prev.shape

        # 1. Affine transformation
        A = np.matmul(x, Wx) + np.matmul(h_prev, Wh) + b

        # 2. slice (4등분)
        f = A[:, :H]
        g = A[: H:2*H]
        i = A[:, 2*H:3*H]
        o = A[:, 3*H:]

        f = sigmoid(f)
        g = np.tanh(g)
        i = sigmoid(i)
        o = sigmoid(o)
        
        c_next = f * c_prev + g * i
        h_next = o * np.tanh(c_next)

        self.cache = (x, h_prev, h_next, i, f, g, o, c_next)
        return h_next, c_next
    
    def backward(self, dh_next, dc_next):
        Wx, Wh, b = self.params
        x, h_prev, h_next, i, f, g, o, c_next = self.cache

        # gate 역전파
        tanh_c_next = np.tanh(c_next)
        ds = dh_next * o * (1 - tanh_c_next**2) + dc_next

        # memory cell 기울기
        dc_prev = ds * f

        # output 게이트
        do = dh_next * tanh_c_next # 
        do *= o * (1 - o)

```

```python
class TimeLSTM:
    def __init__(self, Wx, Wh, b, stateful=False):
        self.params = [Wx, Wh, b]
        self.grads = [np.zeros_like(Wx), np.zeros_like(Wh), np.zeros_like(b)]
        self.layers = None

        self.h, self.c = None, None # Truncated BPTT 할 때
        self.dh = None
        self.stateful = stateful

    def forward(self, xs):
        Wx, Wh, b = self.params
        N, T, D = xs.shape # D는 입력 차원
        H = Wh.shape[0]

        self.layers = []
        # 출력 그릇
        hs = np.empty((N, T, H), dype='f')

        if not self.stateful or self.h is None: # 처음, 무상태, True여야 순전파 계속 전달
            self.h = np.zeros((N, H), dype='f')
        if not self.stateful or self.c is None:
            self.c = np.zeros((N, H), dtype='f')

        # lstm : forward & 인스턴스 변수 layer 추가
        for t in range(T):
            layer = LSTM(*self.params)
            self.h, self.c = layer.forward(xs[:, t, :], self.h, self.c) # h_next, c_next
            hs[:, t, :] = self.h
        
            self.layers.append(layer)

        return hs # 출력
    
    def backward(self, dhs):
        Wx, Wh, b = self.params
        N, T, H = dhs.shape # = hs.shape
        D = Wx.shape[0]

        # 입력 기울기 그릇
        dxs = np.empty((N, T, D), dtype='f')
        dh, dc = 0, 0

        grads = [0, 0, 0] # Truncated BPTT
        for t in reversed(range(T)):
            layer = self.layers[t]
            dx, dh, dc = layer.backward(dhs[:, t, :] + dh, dc) # 분기했던거 합으로
                                                               # (self, dh_next, dc_next) -> (dx, dh_prev, dc_prev)
            dxs[:, t, :] = dx
            for i, grad in enumerate(layer.grads):
                grads[i] += grad # params 순서대로 

        for i, grad in enumerate(grads):
            self.grads[i][...] = grad
        self.dh = dh # dh_prev

        return dxs
    
    def set_state(self, h, c=None):
        self.h, self.c = h, c

    def reset_state(self):
        self.h, self.c = None, None               
```

```python
import sys
sys.path.append('..')
from common.time_layers import *
import pickle

class Rnnlm:
    def __init__(self, vocab_size=10000, wordvec_size=100, hidden_size=100):
        V, D, H = vocab_size, wordvec_size, hidden_size
        rn = np.random.randn

        # 가중치 초기화
        embed_W = (rn(V, D) / 100).astype('f')
        lstm_Wx = (rn(D, 4 * H) / np.sqrt(D)).astype('f') # xavier
        lstm_Wh = (rn(H, 4 * H) / np.sqrt(H)).astype('f')
        lstm_b = np.zeros(4 * H).astype('f')
        affine_W = (rn(H, V) / np.sqrt(H)).astype('f')
        affine_b = np.zeros(V).astype('f')

        # 계층 생성
        self.layers = [
            TimeEmbedding(embed_W),
            TimeLSTM(lstm_Wx, lstm_Wh, lstm_b, stateful=True),
            TimeAffine(affine_W, affine_b)
        ]
        self.loss_layer =TimeSoftmaxWithLoss()
        self.lstm_layer = self.layers[1]

        # 모든 가중치와 기울기를 리스트에 모은다.
        self.params, self.grads = [], []
        for layer in self.layers:
            self.params += layer.params
            self.grads += layer.grads

    def predict(self, xs):
        for layer in self.layers:
            xs = layer.forward(xs)
        return xs
    
    def forward(self, xs, ts):
        score = self.predict(xs)
        loss = self.loss_layer.forward(score, ts)
        return loss

    def backward(self, dout=1):
        dout = self.loss_layer.backward(dout)
        for layer in reversed(self.layers):
            dout = layer.backward(dout)
        return dout
    
    def reset_state(self):
        self.lstm_layer.reset_state()
    
    def save_params(self, file_name='Rnnlm.pkl'):
        with open(file_name, 'wb') as f:
            pickle.dump(self.params, f)

    def load_params(self, file_name='Rnnlm.pkl'):
        with open(file_name, 'rb') as f:
            self.params = pickle.load(f)

```

```python
import sys
sys.path.append('..')
from common.optimizer import SGD
from common.trainer import RnnlmTrainer
from common.util import eval_perplexity
from dataset import ptb
from rnnlm import Rnnlm

# 하이퍼파라미터 설정
batch_size = 20
wordvec_size = 100
hidden_size = 100
time_size = 35
lr = 20.0
max_epoch = 4
max_grad = 0.25 # gradient clipping

# 학습 데이터 읽기 ; 10,000개
corpus, word_to_id, id_to_word = ptb.load_data('train')
corpus_test, _, _ = ptb.load_data('test')
vocab_size = len(word_to_id)
xs = corpus[:-1]
ts = corpus[1:] # 하나 shift

# 모델 생성 및 optmizer, trainer 객체 생성
model = Rnnlm(vocab_size, wordvec_size, hidden_size) # embed (v,w)
optimizer = SGD(lr)
trainer = RnnlmTrainer(model, optimizer) # forward - back - grads 구하기 - 클리핑 - optimizer

# 기울기 클리핑 적용하여 학습 : 4에폭 (1327 * 4)
trainer.fit(xs, ts, max_epoch, batch_size, time_size, max_grad, eval_interval=10)
trainer.plot(ylim=(0, 500))

# 테스트 데이터로 평가
model.reset_state() # hidden state와 memory cell reset
ppl_test = eval_perplexity(model, corpus_test) # 확률의 역수(작을수록), 다음 올 단어의 후보 수
print('테스트 퍼플렉서티: ', ppl_test)

# 매개변수 저장
model.save_params()
```
