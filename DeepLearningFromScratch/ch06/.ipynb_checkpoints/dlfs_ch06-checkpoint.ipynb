```python
class SGD:
    def __init__(self, lr=0.01): # 0.01 / 0.001
        self.lr = lr

    def update(self, params, grads):
        for key in params.key(): # 'W1', 'b1'...
            params[key] -= self.lr * grads[key]
```

```python
class Momentum: 
    def __init__(self, lr=0.01, momentum=0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = None # velocity

    def update(self, params, grads):
        if self.v is None: # no velocity(처음)
            self.v = {}
            for key, val in params.items():
                self.v[key] = np.zeros_like(val) # parameter와 같은 구조 {}

        for key in params.keys():
            self.v[key] = self.momentum*self.v[key] - self.lr*grads[key]
            params[key] += self.v[key]
```

```python
class AdaGrad:
    def __init__(self, lr=0.01):
        self.lr = lr
        self.h = None

    def update(self, params, grads):
        if self.h is None:
            self.h = {}  # grads 누적
            for key, val in params.items():
                self.h[key] = np.zeors_like(val)

        for key in params.keys():
            self.h[key] += grads[key] * grads[key] # ~현재
            params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7)
```

```python
class Adam:                    # 모멘텀(운동량/관성) 계수 (default 값들임)
    def _init__(self, lr=0.001, beta1=0.9, beta2=0.999): 
        self.lr = lr
        self.beta1 = beta1   # 모멘텀(운동량/관성) 계수
        self.beta2 = beta2   # 모멘텀(운동량/관성) 계수
        self.iter = 0   # 몇 번 iteration
        self.m = None  # 1차 관성 속도(momentum)
        self.v = None  # 2차 관성(제곱)(RMSProp)

    def update(self, params, grads):
        # 기존과 동일한 initialization
        if self.m is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)
        # 
        self.iter += 1
        ir_t = self.lr * np.sqrt(1.0 - self.beta2**self.iter) / (1.0 - self.beta1**self.iter)

        for key in params.keys():
            self.m[key] += (1 - self.beta1) * (grads[key] - self.m[key])
            self.v[key] += (1 - self.beta2) * (grads[key]**2 - self.v[key])

            params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-7)  
```

위 코드는 Adam 클래스의 `update` 메서드입니다. 이 메서드는 Adam 최적화 알고리즘을 사용하여 매개변수를 업데이트하는 역할을 합니다.

1. `self.iter += 1`: 반복 횟수(iteration)를 증가시킵니다. 이는 현재까지의 업데이트 횟수를 추적하는 변수입니다.

2. `lr_t = self.lr * np.sqrt(1.0 - self.beta2**self.iter) / (1.0 - self.beta1**self.iter)`: 적응적 학습률(adaptive learning rate)을 계산합니다. Adam 알고리즘에서는 학습률을 반복 횟수에 따라 조정하여 최적화 과정을 안정화시킵니다. `self.lr`은 초기 학습률이며, `self.beta1`과 `self.beta2`는 각각 모멘텀 계수와 RMSProp 계수입니다. 반복 횟수가 증가함에 따라 적응적 학습률이 줄어듭니다.

3. `for key in params.keys():`: 매개변수 딕셔너리의 각 항목에 대해 반복합니다.

4. `self.m[key] += (1 - self.beta1) * (grads[key] - self.m[key])`: 1차 모멘텀(momentum)을 업데이트합니다. 현재 그래디언트와 이전 모멘텀 간의 가중 평균을 계산하여 새로운 모멘텀을 계산합니다.

5. `self.v[key] += (1 - self.beta2) * (grads[key]**2 - self.v[key])`: 2차 모멘텀(RMSProp)을 업데이트합니다. 제곱 그래디언트의 이동 평균을 계산하여 새로운 2차 모멘텀을 계산합니다.

6. `params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-7)`: 매개변수를 업데이트합니다. Adam 알고리즘에서는 모멘텀과 RMSProp을 조합하여 업데이트를 수행합니다. 모멘텀과 RMSProp의 값에 따라 매개변수를 조정하고, 이를 적응적 학습률(`lr_t`)로 조절하여 최적화 과정을 안정화시킵니다. 1e-7을 더하는 것은 0으로 나누는 것을 방지하기 위한 안전 임계값입니다.

```python
import sys, os
sys.path.append(os.pardir)
import numpy as np
from collections import OrderedDict
from common.layers import *
from common.gradient import numerical_gradient

# " 완전연결 다층 신경망 "
class MultiLayerNet: 
    """ 
    parameters  input_size,
                hidden_size_list -> 각 hidden layer의 뉴런 수를 담은 리스트이다. (e.g. [100, 100, 100]）
                output_size,
                activation       -> 'relu'(df) or 'sigmoid' 
                weight_init_std  -> 가중치 표준편차
                                 -> 'relu(df)'/ 'he'로 지정하면, "He 초깃값"(df)
                                 -> 'sigmoid'(tanh) / 'xavier', "Xavier 초깃값" 
                weight_decay_lambda -> L2 Regularization
    """
    def __init__(self, input_size, hidden_size_list, output_size,   
                 activation='relu', weight_init_std='relu', weight_decay_lambda=0): # relu 활성화 & He 초깃값
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size_list = hidden_size_list
        self.hidden_layer_num = len(hidden_size_list)  # 새로운 인스턴스 변수-은닉층개수
        self.weight_decay_lambda = weight_decay_lambda
        self.params = {}

        # 가중치 초기화 (__init_weight 메서드 밑에)
        self.__init_weight(weight_init_std)
        
        # 계층 생성
        activation_layer = {'sigmoid' : Sigmoid, 'relu' : Relu} # 함수(class) 매핑(layers.py)
        self.layers = OrderedDict()
        for idx in range(1, self.hidden_layer_num + 1): # 은닉층 개수만큼 affine&act
            self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)],
                                                     self.params['b' + str(idx)])                                                                          
            self.layers['Activation_function' + str(idx)] = activation_layer[activation]() # Sigmoid()/Relu()
                                                                         # 생성자,paramemter
        # to the last layer!
        idx = self.hidden_layer_num + 1
        self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)],
            self.params['b' +  str(idx)])
        self.last_layer = SoftmaxWithLoss()

    def __init_weight(self, weight_init_std):
        # relu/he - he
        # sigmoid/xavier - xavier
        all_size_list = [self.input_size] + self.hidden_size_list + [self.output_size] # [784, 100, 100, 100, 10]
        for idx in range(1, len(all_size_list)): # 1~#모든계층
            scale = weight_init_std
            if str(weight_init_std).lower() in ('relu', 'he'): # Relu -> He 초깃값
                scale = np.sqrt(2.0 / all_size_list[idx - 1]) # 0(input)~...
            elif str(weight_init_std).lower() in ('sigmoid', 'xavier'):
                scale = np.sqrt(1.0 / all_size_list[idx - 1])    # (input, output)
            self.params['W' + str(idx)] = scale * np.random.randn(all_size_list[idx-1], all_size_list[idx])
            self.params['b' + str(idx)] = np.zeros(all_size_list[idx]) #(output,)

    def predict(self, x):
        for layer in self.layers.values():
            x = layer.forward(x) # 계층 순서대로~학습

        return x
        
    def loss(self, x, t):
        y = self.predict(x)

        # 가중치 감소
        weight_decay = 0
        for idx in range(1, self.hidden_layer_num + 2):
            W = self.params['W' + str(idx)]
            weight_decay += 0.5 * self.weight_decay_lambda * np.sum(W ** 2)
        return self.last_layer.forward(y, t) + weight_decay # softmaxwithloss

    def accuracy(self, x, t):
        y = self.predict(x)
        y = np.argmax(y, axis=1)  # logit
        if t.ndim != 1 : t = np.argmax(t, axis=1) # 원핫 -> label 추출

        accuracy = np.sum(y == t) / float(x.shape[0])
        return accuracy

    def numerical_gradient(self, x, t):
        loss_W = lambda W: self.loss(x, t)

        grads = {}
        for idx in range(1, self.hidden_layer_num + 2): # 끝까지 (last-affine)
            grads['W' + str(idx)] = numerical_gradient(loss_W, self.params['W' + str(idx)])
            grads['b' + str(idx)] = numerical_gradient(loss_W, self.params['b' + str(idx)])

        return grads

    def gradient(self, x, t):
        # forward
        self.loss(x, t)

        # backward
        dout = 1
        dout = self.last_layers.backward(dout)

        layers = list(self.layers_values())
        layers.reverse()
        for layer in layers:
            dout = layer.backward(dout)

        # 기울기 결과 저장
        grads = {}
        for idx in range(1, self.hidden_layer_num + 2):
            grads['W' + str(idx)] = self.layers['Affine' + str(idx)].dW + self.weight_decay_lambda * self.layers['Affine' + str(idx)].W
            grads['b' + str(idx)] = self.layers['Affine' + str(idx)].db

        return grads



# 0. MNIST 데이터 읽기==========
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)  # 원핫레이블x
train_size = x_train.shape[0] # 60000
batch_size = 128 # 안나눠 떨어짐 (1epoch-468.75)
max_iterations = 2000 

hey = MultiLayerNet(input_size=784, hidden_size_list=[100, 100, 100, 100],
        output_size=10)
hey.loss(x_train, t_train)
```

```python
# coding: utf-8
import os
import sys
sys.path.append(os.pardir)  # 부모 디렉터리의 파일을 가져올 수 있도록 설정
import matplotlib.pyplot as plt
from dataset.mnist import load_mnist
from common.util import smooth_curve
from common.multi_layer_net import MultiLayerNet
from common.optimizer import *


# 0. MNIST 데이터 읽기==========
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)
train_size = x_train.shape[0]
batch_size = 128
max_iterations = 2000


# 1. 실험용 설정==========
optimizers = {}
optimizers['SGD'] = SGD()
optimizers['Momentum'] = Momentum()
optimizers['AdaGrad'] = AdaGrad()
optimizers['Adam'] = Adam()
#optimizers['RMSprop'] = RMSprop()

networks = {}
train_loss = {}
for key in optimizers.keys():
    networks[key] = MultiLayerNet(
        input_size=784, hidden_size_list=[100, 100, 100, 100],
        output_size=10)
    train_loss[key] = []    


# 2. 훈련 시작==========
for i in range(max_iterations):
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]
    
    for key in optimizers.keys():
        grads = networks[key].gradient(x_batch, t_batch)
        optimizers[key].update(networks[key].params, grads)
    
        loss = networks[key].loss(x_batch, t_batch)
        train_loss[key].append(loss)
    
    if i % 100 == 0:
        print( "===========" + "iteration:" + str(i) + "===========")
        for key in optimizers.keys():
            loss = networks[key].loss(x_batch, t_batch)
            print(key + ":" + str(loss))


# 3. 그래프 그리기==========
# markers = {"SGD": "o", "Momentum": "x", "AdaGrad": "s", "Adam": "D"}
# x = np.arange(max_iterations)
# for key in optimizers.keys():
#     plt.plot(x, smooth_curve(train_loss[key]), marker=markers[key], markevery=100, label=key)
# plt.xlabel("iterations")
# plt.ylabel("loss")
# plt.ylim(0, 1)
# plt.legend()
# plt.show()
```

```python
def smooth_curve(x):
    # 1. 윈도우 사이즈 설정
    window_len = 11
    # 2. 앞 뒤 윈도우로 자른 것과 원래 concat -> "대칭 확장"
    s = np.r_[x[window_len-1:0:-1], x, x[-1:-window_len:-1]]
    # 3. 카이저 윈도우 
    w = np.kaiser(window_len, 2)
    # 4. 두 1차원 시퀀스의 이산 선형 컨볼루션, 데이터를 필터링하고 평활화할 수 있는 이동 평균을 생성
    y = np.convolve(w/w.sum(), s, mode='valid') # valid 입출력신호 사이즈 맞게
    # 5. 윈도우 크기 11, 가장자리 없애주기
    return y[5:len(y)-5]
    
"""
신호 처리 및 데이터 분석에서 일반적인 윈도우 기능
Kaiser 창은 스펙트럼 누출 인공물과 신호 누출을 크게 줄여 데이터 분석 방법의 정밀도와 신뢰성을 향상시킵니다.
스펙트럼 분석 중 관심 신호가 인접한 주파수 빈에 분산될 때 발생하는 누출을 줄이기 위한 것입니다.
numpy.kaiser(M, beta, sym=True) 여기서 M은 창의 길이를 나타내고, beta는 창의 형태를 결정하며, Sym은 창이 대칭이어야 하는지 여부를 지정합니다.
'M' - 창의 길이. 일반적으로 0보다 큰 양의 정수입니다.
'베타' - 카이저 창의 모양 매개변수입니다. 이는 사이드로브 수준과 메인 로브 폭 사이의 절충안을 조절합니다. 
레벨이 낮을수록 메인 로브가 넓어지고 사이드 로브가 늘어나는 반면, 
베타가 커지면 창이 좁아지므로 점점 더 좁아지는 스파이크를 샘플링할 수 있을 만큼 샘플 수가 커야 합니다. 일반적으로 양의 실수입니다.
'sym' - 창이 대칭이어야 하는지 여부를 제어하는 비필수 매개변수입니다. True로 설정하면 결과 창은 대칭이 됩니다.

`return y[5:len(y)-5]`는 함수의 출력을 조정하기 위한 부분입니다.
위 코드에서 `np.convolve` 함수를 사용하여 이동 평균을 생성한 후에는 윈도우의 크기에 따라 출력의 길이가 입력보다 커집니다. 
이는 이동 평균을 계산할 때 데이터의 가장자리 부분에서는 완전한 평균을 구하지 못하기 때문에 발생합니다. 
따라서 이동 평균의 가장자리 부분은 부적절한 값이 될 수 있습니다.
즉, `y[5:len(y)-5]`는 이러한 부적절한 가장자리 부분을 제거하여 유효한 출력만을 반환
이 부분은 앞뒤로 5개의 데이터를 제외하고 출력을 선택하도록 하는 것으로, 이동 평균의 중앙에 해당하는 부분만을 반환
"""
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    if x.ndim == 2:
        x = x.T
        x = x - np.max(x, axis=0)
        y = np.exp(x) / np.sum(np.exp(x), axis=0)
        return y.T 
    
    x = x - np.max(x) # 오버플로 대책
    return np.exp(x) / np.sum(np.exp(x))

def cross_entropy_error(y, t):
    if y.ndim == 1:
        t = t.reshape(1, t.size)
        y = y.reshape(1, y.size)

    if t.size == y.size:
        t = t.argmax(axis=1)
             
    batch_size = y.shape[0]
    return -np.sum(np.log(y[np.arange(batch_size), t] + 1e-7)) / batch_size

class TwoLayerNet:
    def __init__(self, input_size, hidden_size, output_size, weight_init=0.1):
        self.params = {}
        self.params['W1'] = weight_init * np.ones((input_size, hidden_size))
        self.params['b1'] = np.zeros(hidden_size)
        self.params['W2'] = weight_init * np.ones((hidden_size, output_size))
        self.params['b2'] = np.zeros(output_size)
        
    def predict(self, x):
        W1, W2 = self.params['W1'], self.params['W2']
        b1, b2 = self.params['b1'], self.params['b2']
        
        a1 = np.dot(x, W1) + b1
        z1 = sigmoid(a1)
        a2 = np.dot(z1, W2) + b2
        y = softmax(a2)
        
        return y
        
    def loss(self, x, t):
        y = self.predict(x)
        
        return cross_entropy_error(y, t)
    
    def accuracy(self, x, t):
        y = self.predict(x)
        y = np.argmax(y, axis=1)
        t = np.argmax(t, axis=1)
        
        accuracy = np.sum(y == t) / float(x.shape[0])
        return accuracy
    
    def numerical_gradient(self, x, t):
        loss_W = lambda W: self.loss(x, t)
        
        grads = {}
        grads['W1'] = numerical_gradient(loss_W, self.params['W1'])
        grads['b1'] = numerical_gradient(loss_W, self.params['b1'])
        grads['W2'] = numerical_gradient(loss_W, self.params['W2'])
        grads['b2'] = numerical_gradient(loss_W, self.params['b2'])
        
        return grads
```

```python
import os, sys
sys.path.append(os.pardir) 
import numpy as np
import matplotlib.pyplot as plt
from dataset.mnist import load_mnist
from common.util import smooth_curve 
from common.multi_layer_net import MultiLayerNet 
from common.optimizer import SGD


# 0. MNIST 데이터 읽기==========
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)

train_size = x_train.shape[0]
batch_size = 128
max_iterations = 2000


# 1. 실험용 설정==========
weight_init_types = {'std=0.01': 0.01, 'Xavier': 'sigmoid', 'He': 'relu'}
optimizer = SGD(lr=0.01)

networks = {}
train_loss = {}
for key, weight_type in weight_init_types.items():
    networks[key] = MultiLayerNet(input_size=784, hidden_size_list=[100, 100, 100, 100],
                                  output_size=10, weight_init_std=weight_type)
    train_loss[key] = []


# 2. 훈련 시작==========
for i in range(max_iterations):
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]
    
    for key in weight_init_types.keys():
        grads = networks[key].gradient(x_batch, t_batch)
        optimizer.update(networks[key].params, grads)
    
        loss = networks[key].loss(x_batch, t_batch)
        train_loss[key].append(loss)
    
    if i % 100 == 0:
        print("===========" + "iteration:" + str(i) + "===========")
        for key in weight_init_types.keys():
            loss = networks[key].loss(x_batch, t_batch)
            print(key + ":" + str(loss))


# 3. 그래프 그리기==========
markers = {'std=0.01': 'o', 'Xavier': 's', 'He': 'D'}
x = np.arange(max_iterations)
for key in weight_init_types.keys():
    plt.plot(x, smooth_curve(train_loss[key]), marker=markers[key], markevery=100, label=key)
plt.xlabel("iterations")
plt.ylabel("loss")
plt.ylim(0, 2.5)
plt.legend()
plt.show()
```

```python
import sys, os
sys.path.append(os.pardir)  # 부모 디렉터리의 파일을 가져올 수 있도록 설정
import numpy as np
import matplotlib.pyplot as plt
from dataset.mnist import load_mnist
from common.multi_layer_net_extend import MultiLayerNetExtend
from common.optimizer import SGD, Adam

(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True) # 픽셀0~1

#학습 데이터를 줄임
x_train = x_train[:1000]
t_train = t_train[:1000]

max_epochs = 20
train_size = x_train.shape[0] # 1000
batch_size = 100 # epoch 10
learning_rate = 0.01

def __train(weight_init_std):
    # if batch normalization
    bn_network = MultiLayerNetExtend(input_size=784, hidden_size_list=[100, 100, 100, 100, 100], output_size=10,  # 6층
                                     weight_init_std=weight_init_std, use_batchnorm=True)
    # ~bn
    network = MultiLayerNetExtend(input_size=784, hidden_size_list=[100, 100, 100, 100, 100], output_size=10,
                                weight_init_std=weight_init_std)
    optimizer = SGD(lr=learning_rate)

    train_acc_list = []
    bn_train_acc_list = []

    iter_per_epoch = max(train_size / batch_size, 1) # 10
    epoch_cnt=0

    for i in range(1000000000):
        batch_mask = np.random.randn(train_size, batch_size)
        x_batch = x_train[batch_mask]
        t_batch = t_train[batch_mask]

        for _network in (bn_network, network): # 2가지 경우
            grads = _network.gradient(x_batch, t_batch)
            optimizer.update(_network.params, grads)

        if i % iter_per_epoch == 0:
            train_acc = network.accuracy(x_train, t_train)
            bn_train_acc = bn_network.accuracy(x_train, t_train)
            train_acc_list.append(train_acc)
            bn_train_acc_list.append(bn_train_acc)

            print("epoch:" + str(epoch_cnt) + " | " + str(train_acc) + " - " + str(bn_train_acc))

            epoch_cnt += 1
            if epoch_cnt >= max_epochs: # epoch0~19(20번)
                break

    return train_acc_list, bn_train_acc_list

# 학습 & 그래프 그리기
weight_scale_list = np.logspace(0, -4, num=16) # 0~-4까지의 범위에서 로그로 분할한 16개의 값을 생성
# log scale 등간격인 행 벡터를 생성 (일정 간격 벡터값)(e)

x = np.arange(max_epochs) # array([0,...,19])

for i, w in enumerate(weight_scale_list): # idx,value
    print( "============== " + str(i+1) + "/16" + " ==============")
    train_acc_list, bn_train_acc_list = __train(w) # 학습 시작!

    plt.subplot(4, 4, i+1) # r/c/idx1부터
    plt.title("W:" + str(w))
    if i == 15: # 최종
        plt.plot(x, bn_train_acc_list, label='Batch Normalization', markevery=2)
        plt.plot(x, train_acc_list, linestyle = "--", label='Normal(without BatchNorm)', markevery=2)
    else:
        plt.plot(x, bn_train_acc_list, markevery=2)
        plt.plot(x, train_acc_list, linestyle="--", markevery=2)

    plt.ylim(0, 1.0)
    if i % 4: # 0이 아니면 즉, 각 row에 첫 그래프 제외하고
        plt.yticks([])
    else:
        plt.ylabel("accuaracy")
        
    if i < 12: # 13부터 즉, 마지막 라인은 x 눈금 표시
        plt.xticks([])
    else:
        plt.xlabel("epochs")
    plt.legend(loc='lower right') # 최종그래프

plt.show()
```

```python
import sys, os
sys.path.append(os.pardir) 
import numpy as np
from collections import OrderedDict
from common.layers import *
from common.gradient import numerical_gradient

# 가중치 감소, 드롭아웃, 배치 정규화
# 추가된 params - use_dropout : 드롭아웃 사용 여부 / dropout_ration : 드롭아웃 비율 /
#                use_batchNorm : 배치 정규화 사용 여부

class MultiLayerNetExtend:
    def __init__(self, input_size, hidden_size_list, output_size,
                 activation='relu', weight_init_std='relu', weight_decay_lambda=0, 
                 use_dropout=False, dropout_ration=0.5, use_batchnorm=False):
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size_list = hidden_size_list
        self.hidden_layer_num = len(hidden_size_list)
        # new!
        self.use_dropout = use_dropout
        self.weight_decay_lambda = weight_decay_lambda
        self.use_batchnorm = use_batchnorm
        self.params = {}
        # 가중치 초기화
        self.__init_weight(weight_init_std)

        # 계층 생성
        activation_layer = {'sigmoid' : Sigmoid, 'relu' : Relu}
        self.layers = OrderedDict()
        for idx in range(1, self.hidden_layer_num+1): # 1~hidden개수까지
            self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)],
                                                      self.params['b' + str(idx)])
            # activation(hiddeen) 전 batchnorm
            if self.use_batchnorm: # gamma=1, beta=0부터 시작 -> 학습 조정
                self.params['gamma' + str(idx)] = np.ones(hidden_size_list[idx-1])
                self.params['beta' + str(idx)] = np.zeros(hidden_size_list[idx-1])
                self.layers['BatchNorm' + str(idx)] = BatchNormalization(self.params['gamma' + str(idx)], self.params['beta' + str(idx)])

            if self.use_dropout:
                 self.layers['Dropout' + str(idx)] = Dropout(dropout_ration)

        idx = self.hidden_layer_num + 1 # to the last
        self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)], self.params['b' + str(idx)])
        self.last_layer = SoftmaxWithLoss()

    #def __init_weight(self, weight_init_std): 동일

    def predict(self, x, train_flg=False):
        for key, layer in self.layers.items():
            if "Dropout" in key or "BatchNorm" in key:
                x = layer.forward(x, train_flg)   # train모드 일때만 작동
            else: # 동일
                x = layer.forward(x)

        return x

    def loss(self, x, t, train_flg=False):
       y = self.predict(x, train_flg)
        
        # 동일... 가중치 감소 다음절

    def accuracy(self, X, T):
        Y = self.predict(X, train_flg=False) # accuracy땐 무조건 false
        # 동일...

    def numerical_gradient(self, X, T):
        loss_W = lambda W : self.loss(X, T, train_flg=True) # train
        
        grads = {}
        for idx in range(1, self.hidden_layer_num+2): # 끝
            grads['W' + str(idx)] = numerical_gradient(loss_W, self.params['W' + str(idx)])
            grads['b' + str(idx)] = numerical_gradient(loss_W, self.params['b' + str(idx)])
            
            if self.use_batchnorm and idx != self.hidden_layer_num+1: # 마지막 빼고!
                grads['gamma' + str(idx)] = numerical_gradient(loss_W, self.params['gamma' + str(idx)])
                grads['beta' + str(idx)] = numerical_gradient(loss_W, self.params['beta' + str(idx)])
                
        return grads

    def gradient(self, x, t):
        # forward
        self.loss(x, t, train_flg=True)
        # backward
        # 동일..

        # 결과 저장
        grads = {}
        for idx in range(1, self.hidden_layer_num+2):
            grads['W' + str(idx)] = self.layers['Affine' + str(idx)].dW + self.weight_decay_lambda * self.params['W' + str(idx)]
            grads['b' + str(idx)] = self.layers['Affine' + str(idx)].db

            if self.use_batchnorm and idx != self.hidden_layer_num+1:
                grads['gamma' + str(idx)] = self.layers['BatchNorm' + str(idx)].dgamma
                grads['beta' + str(idx)] = self.layers['BatchNorm' + str(idx)].dbeta

        return grads        
```

```python
class Dropout:
    def __init__(self, dropout_ratio=0.5):
        # 확률을 이용해 hidden 뉴런 끊어버릴 것
        self.dropout_ratio = dropout_ratio
        self.mask = None

    def forward(self, x, train_flg=True):
        if train_flg:
            # True/False mask mat
            self.mask = np.random.rand(*x.shape) > self.dropout_ratio
            return x * self.mask # 끊어서 학습하겠다.

        else: # test
            # 스케일을 학습시와 맞춰주기 위해 삭제 안 한 비율
            return x * (1.0 - self.dropout_ratio)
            """
            (1.0 - self.dropout_ratio)는 활성화된 뉴런들의 비율을 의미
            학습 시에는 일부 뉴런을 끄기 때문에 비활성화된 뉴런의 비율인 dropout_ratio를 빼줘야한다. 
            그러면 남은 활성화된 뉴런들의 비율이 되는 것이죠.
            평균 내는 효과? 근데 실제로 그냥 곱하지 않고 쓴다함
            """
    def backward(self, dout):
        # 순전파 때처럼 죽이기 (relu: dout[mask] = 0)
        return dout * self.mask
```

```python
import sys, os
sys.path.append(os.pardir)
import numpy as np
from collections import OrderedDict
from common.layers import *
from common.gradient import numerical_gradient

class MultiLayerNetExtend:
    def __init__(self, input_size, hidden_size, output_size,
                 activation='relu', weight_init_std='relu', weight_decay_lambda=0,
                 use_dropout=False, dropout_ration=0.5, use_batchnorm=False):
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size_list = hidden_size_list
        self.hidden_layer_num = len(hidden_size_list)
        self.use_dropout = use_dropout
        self.weight_decay_lambda = weight_decay_lambda
        self.use_batchnorm = use_batchnorm
        self.params = {}

        # 가중치 초기화
        self.__init_weight(weight_init_std)

        # 계층 생성
        activation_layer = {'sigmoid' : Sigmoid, 'relu' : Relu}
        self.layers = OrderedDict()
        for idx in range(1, self.hidden_layer_num + 1):
            self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)],
                                                      self.params['b' + str(idx)])
            if self.use_batchnorm:
                self.params['gamma' + str(idx)] = np.ones(hidden_size_list[idx-1])
                self.parmas['beta' + str(idx)] = np.zeros(hidden_size_list[idx-1])
                self.layers['BatchNorm' + str(idx)] 
                = BatchNormalization(self.params['gamma' + str(idx)], self.params['beta' + str(idx)])

            self.layers['Activation_function' + str(idx)] = actvation_layer[activation]()

            # dropout은 activation 다음! 출력 안뱉도록
            if self.use_dropout:
                self.layers['Dropout' + str(idx)] = Dropout(dropout_ration)

        idx = self.hidden_layer_num + 1
        self.layers['Affine' + str(idx)] = Affine(self.params['W' + str(idx)], self.params['b' + str(idx)])
        self.last_layer = SoftmaxWithLoss()

    # 가중치 초기화 fun

    def predict(self, x, train_flg=False):
        for key, layer in self.layer.items():
            if "Dropout" in key or "BatchNorm" in key:
                x = layer.forward(x, train_flg)
            else:
                x = layer.forward(x)
        return x

    # loss/gradient 동일...
```

```python
def shuffle_dataset(x, t):
    # permutation : 순서가 부여된 임의의 집합을 다른 순서로 뒤섞는 연산
    permutation = np.random.permutation(x.shape[0])
    # np.random.shuffle(x)는 x 자체를 자동으로 바꿔버림
    x = x[permutation,:] if x.ndim==2 else x[permutation,:,:,:]
    t = t[permutation]

    return x, t
```

```python
class BatchNormalization:
    def __init__(self, gamma, beta, momentum=0.9, running_mean=None, running_var=None):
        self.gamma = gamma
        self.beta = beta
        self.momentum = momentum
        self.input_shape = None # 합성곱 계층은 4차원, 완전연결 계층은 2차원

        # 시험할 때 사용할 평균과 분산
        self.running_mean = running_mean
        self.running_var = running_var
        """
- 학습 단계 (training phase): 각 미니배치에서 feature의 평균과 분산을 계산하고, 이를 사용해서 해당 배치를 정규화 합니다. 
                             이 과정에서 추가적인 스케일 파라미터(γ)와 이동 파라미터(β)가 적용됩니다. 이 파라미터들은 학습 과정에서 최적화 됩니다. 
                             동시에, 미니배치의 평균과 분산에 대한 이동 평균(moving average)가 계산되어 저장됩니다. 이 값들은 추론 단계에서 사용됩니다.
                                
- 추론 단계 (intference phase): 학습 단계에서 계산된 이동 평균을 사용하여 배치를 정규화합니다. 
                                이 때는 새로운 데이터에 대해 예측을 수행하기 때문에, 
                                미니배치의 평균과 분산을 계산하는 대신 학습 단계에서 저장된 이동 평균을 사용합니다.
                                이렇게 추론단계에서, 훈련 단계에서 계산된 평균과 분산의 이동 평균을 사용하여 정규화를 하는데, 
                                그 이유는, 보통 하나의 샘플 또는 매우 작은 배치의 데이터에 대한 예측을 수행하기 때문에, 
                                이 때 미니배치의 평균과 분산을 계산하는 것이 불가능하거나 불안정할 수 있기 때문입니다. 
                                따라서, 훈련 과정에서 모든 미니배치 통계치를 고려한 이동 평균을 사용하여 이 문제를 해결합니다. 
                                -> 애초에 SGD에서 미니배치의 크기가 1이라고 할 때 이는 분산이 0이 될 뿐더러 평균도 0이 됩니다.
"""

        # backward시 사용할 중간 데이터
        self.batch_size = None
        self.xc = None
        self.std = None
        self.dgamma = None
        self.dbeta = None

    def forward(self, x, train_flg=True):
        self.input_shape = x.shape
        if x.ndim != 2: # 합성곱 게층에서 들어온 input flatten
            N, C, H, W = x.shape
            x = x.shape(N, -1)

        out = self.__forward(x, train_flg)

        return out.reshape(*self.input_shape) # 다시 복원

    def __forward(self, x, train_flg):
        # running_ 초기화
        if self.running_mean is None:
            N, D = x.shape # (N, -1)
            self.running_mean = np.zeros(D)
            self.running_var = np.zeros(D)

        if train_flg: # normalizatoin
            mu = x.mean(axis=0) # col에 대한!!
            xc = x - mu
            var = np.mean(xc**2, axis=0) 
            std = np.sqrt(var + 10e-7)
            xn = xc / std

            self.batch_size = x.shape[0]
            self.xc = xc
            self.xn = xn
            self.std = std
            # 매 배치마다 구해지는 mu와 이전에 구해졌던 running_mean을 내분해서 새로운
            # running_mean 갱신!
            self.running_mean = self.momentum * self.running_mean + (1-self.momentum) * mu
            self.running_var = self.momentum * self.running_var + (1-self.momentum) * var 

        else: # 추론 시
            xc = x - self.running_mean
            xn = xc / ((np.squrt(self.running_var + 10e-7)))

        out = self.gamma * xn + self.beta
        return out

    def backward(self, dout):
        if dout.ndim != 2:
            N, C, H, W = dout.shape
            dout = dout.reshape(N, -1) # flatten

        dx = self.__backward(dout)

        dx = dx.reshape(*self.input_shape)
        return dx

    def __backward(self, dout):
        dbeta = dout.sum(axis=0)
        dgamma = np.sum(self.xn * dout, axis=0)
        dxn = self.gamma * dout
        dxc = dxn / self.std
        dstd = -np.sum((dxn * self.xc) / (self.std * self.std), axis=0)
        dvar = 0.5 * dstd / self.std
        # 위쪽에서 구한 dxc와 아래의 dxc가 합류를 하기 때문에 더함
        dxc += (2.0 / self.batch_size) * self.xc * dvar
        dmu = np.sum(dxc, axis=0)
        dx = dxc - dmu / self.batch_size
        
        self.dgamma = dgamma
        self.dbeta = dbeta
        
        return dx
```
