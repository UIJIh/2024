```python
# coding=utf8
# REST API 호출에 필요한 라이브러리
import requests
import json

# [내 애플리케이션] > [앱 키] 에서 확인한 REST API 키 값 입력
REST_API_KEY = 'e2eae1a0d18e30595c7434bc271c8c72'

# KoGPT API 호출을 위한 메서드 선언
# 각 파라미터 기본값으로 설정
def kogpt_api(prompt, max_tokens = 1, temperature = 1.0, top_p = 1.0, n = 1):
    r = requests.post(
        'https://api.kakaobrain.com/v1/inference/kogpt/generation',
        json = {
            'prompt': prompt,
            'max_tokens': max_tokens,
            'temperature': temperature,
            'top_p': top_p,
            'n': n
        },
        headers = {
            'Authorization': 'KakaoAK ' + REST_API_KEY,
            'Content-Type': 'application/json'
        }
    )
    # 응답 JSON 형식으로 변환
    response = json.loads(r.content)
    return response
```

```python
prompt='''정보:아래 발언에 대한 귀하의 동의/비동의 정도를 짧게 말해주십시오.
Q:안녕하세요 반갑습니다. 발언에 대해 "1. 매우 동의하지 않음 / 2. 동의하지 않음 / 3. 중립 / 4. 동의함 / 5. 매우 동의함." 중 하나로 대답해주세요. 간단한 이유도 말해주세요.
A:안녕하세요. 알겠습니다. 
Q:"학교 교육의 주요 기능은 미래 세대가 일자리를 찾을 수 있도록 준비시키는 것이어야 합니다." 어떻게 생각하시나요?  이 발언에 대해 1. 매우 동의하지 않음 / 2. 동의하지 않음 / 3. 중립 / 4. 동의함 / 5. 매우 동의함 중 하나를 선택한다면요? 
A:'''
for i in range(10):
    response = kogpt_api(prompt, max_tokens=70, temperature=0.1, top_p=0.85)
    print(i, response['generations'][0]['text'], "\n")
    # generations(키) 리스트에서 첫 번째 항목에서 text 키 값만
```
