# Token?

A word-based tokenization algorithm will break the sentence into words. The most common one is splitting based on space.

[“Let”, “us”, “learn”, “tokenization.”]

A subword-based tokenization algorithm will break the sentence into subwords.

[“Let”, “us”, “learn”, “token”, “ization.”]

A character-based tokenization algorithm will break the sentence into characters.

[“L”, “e”, “t”, “u”, “s”, “l”, “e”, “a”, “r”, “n”, “t”, “o”, “k”, “e”, “n”, “i”, “z”, “a”, “t”, “i”, “o”, “n”, “.”]

```python
with open('wizard_of_oz.txt', 'r', encoding='utf-8') as f:
    text = f.read()
print(len(text)) # 단어 수
print(text[:200])
print(len(text[:200]))
print()
# tokens
chars = sorted(set(text))
print(chars)
print(len(chars))
```

```python
string_to_int = { ch:i for i,ch in enumerate(chars) } # key:value
int_to_string = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [string_to_int[c] for c in s]
decode = lambda l: ''.join([int_to_string[i] for i in l]) # value는 ch고, join해서 word
encoded_hello = encode('hello')
decoded_hello = decode(encoded_hello)
print(encoded_hello, '\n', decoded_hello)
```

# torch 쓸거임

```python
import torch
data = torch.tensor(encode(text), dtype=torch.long) # a super long sequence
print(data[:100])
```

# traing set / val 으로 분리하기

```python
n = int(0.8*len(data))
train_data = data[:n]
val_data = data[n:]
```

# Bigram Language Model에서 target 나누는 법
offset 하나 shift해서!


```python
block_size = 8

x = train_data[:block_size]
y = train_data[1:block_size+1]

for t in range(block_size):
    context = x[:t+1]
    target = y[t]
    print('when input is', context, 'target is', target)
```
